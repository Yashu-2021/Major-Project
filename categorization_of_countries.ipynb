```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
# Replace 'your_dataset.csv' with the actual file path
df = pd.read_csv('your_dataset.csv')

# Select data for the specified indicators
climate_land_data = df[df['Indicator Code'].isin(selected_indicators)]

# Group data by country and indicator for plotting
grouped_data = climate_land_data.groupby(['Country Name', 'Indicator Code']).mean().reset_index()

# Plot time series trends for each indicator across countries
plt.figure(figsize=(15, 8))
for indicator_code, indicator_data in grouped_data.groupby('Indicator Code'):
    for index, row in indicator_data.iterrows():
        values = row.loc['1960':'1967']
        plt.plot(values, label=f"{row['Country Name']} - {indicator_code}")

plt.title('Trends for Selected Climate Change and Land Degradation Indicators')
plt.xlabel('Year')
plt.ylabel('Indicator Value')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Country - Indicator Code')
plt.show()
```

```python
columns_to_drop = ['Country Name', 'Indicator Name', 'Indicator Code']
filtered_df4.drop(columns=columns_to_drop, inplace=True)

# Display the modified DataFrame
print(filtered_df4)
```

```python
# Convert numeric columns to numeric type, replacing non-numeric values with NaN
filtered_df4.iloc[:, 1:-1] = filtered_df4.iloc[:, 1:-1].apply(pd.to_numeric, errors='coerce')

# Create a new column 'Combined' with the median of values from '1960' to '1965'
filtered_df4['median_aff'] = filtered_df4.iloc[:, 1:-1].median(axis=1)


# Display the resulting DataFrame with the new 'Combined' column
print(filtered_df4[['Country Code', 'median_aff']])
```

```python
low_threshold = filtered_df4['median_aff'].quantile(0.33)
high_threshold = filtered_df4['median_aff'].quantile(0.66)

# Categorize based on thresholds
filtered_df4['Category_median_aff'] = pd.cut(filtered_df4['median_aff'], bins=[-float('inf'), low_threshold, high_threshold, float('inf')],
                               labels=['Low', 'Moderate', 'High'])

# Display the result
print(filtered_df4[['Country Code', 'median_aff', 'Category_median_aff']])
```
